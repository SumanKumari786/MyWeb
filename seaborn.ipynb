# Different types of ghaph making methods

#  distplot()

```python
print(sas.distplot(tips['total_bill']))
```

### distplot() with kde attribute that is used to remove curve

```python
print(sas.distplot(tips['total_bill'],kde=False))
```

### distplot() with bins attribute that is used to divide curves

```python
print(sas.distplot(tips['total_bill'],kde=False,bins=40))
```

# jointplot(x-axis, y-axis, data)

```python
print(sas.jointplot(x="total_bill",y="tip",data=tips))
```

### jointplot() with kind="kde" attribute that is used to make curve

```python
print(sas.jointplot(x="total_bill",y="tip",data=tips,kind="kde"))
```

### jointplot() with kind="hex" attribute that is used to make hexagonals

```python
print(sas.jointplot(x="total_bill",y="tip",data=tips,kind="hex"))
```

# pairplot()

```python
print(sas.pairplot(tips))
```

### pairplot() with hue="sex" attribute that provide segregation & different colors to graph

```python
print(sas.pairplot(tips,hue="sex"))
```

### pairplot() with hue="day" attribute that provide segregation & different colors to graph

```python
print(sas.pairplot(tips,hue="day"))
```

# rugplot()

```python
print(sas.rugplot(tips["total_bill"]))
```
